```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('data/clean_data.csv')
```

Create new columns of binned Median Age and binned Overall hospital rating

```python
bins=[0,30,40,50,60,100]
df['Binned Median Age'] = pd.cut(df['Median Age'], bins)

bins2=[0,3,5]
df['Binned Hospital overall rating'] = pd.cut(df['Hospital overall rating'], bins2)
```

Create dummies for Binned Median Age

```python
dummy = pd.get_dummies(df['Binned Median Age'])
df = pd.concat([df,dummy], axis=1)
df.drop(columns=['Binned Median Age'], inplace=True)
df.info()
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(columns=['Hospital overall rating', 'Mortality national comparison', 'Readmission national comparison', 'Patient experience national comparison', 'Effectiveness of care national comparison', 'Timeliness of care national comparison', 'Median Time Spent in Emergency Department Before Leaving', 'Median Time Spent in Emergency Department After Doctor Decides to Admit as Inpatient'])
y = df['Hospital overall rating']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=1)
```
