```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv('~/Desktop/Springboard/CapstoneTwo/data/clean_data.csv')
```

```python
df.info()
```

```python
sns.swarmplot(data=df,
            x='Hospital overall rating',
            y='Median Age')
plt.show()
```

```python
sns.boxplot(data=df,
            x='Hospital overall rating',
            y='Median Age')
plt.show()
```

It appears as though an increase of median age is positively correlated with an increase in hospital overall rating

```python
sns.swarmplot(data=df,
            x='Hospital overall rating',
            y='Median Household Income')
plt.show()
```

```python
sns.boxplot(data=df,
            x='Hospital overall rating',
            y='Median Household Income')
plt.show()
```

We see a similar trend with median household income, although the increase in median household income is much more subtle from hospital ratings of 1 to 4 and then more dramatic from 4 to 5

```python
sns.boxplot(data=df,
            x='Mortality national comparison',
            y='Median Age')
plt.show()
```

```python
sns.boxplot(data=df,
            x='Mortality national comparison',
            y='Median Household Income')
plt.show()
```

Doesn't seem to be much of a correlation between median age and mortality rate, but seems like there is between income and mortality rate.

```python
sns.set(font_scale=3)
_ = plt.figure(figsize=(50,30))
sns.heatmap(df.corr(), annot=True, annot_kws={"size": 30});
```

I'm interested in correlations in the bottom left of the heatmap. There don't seem to be a ton of strong correlations, positive or negative. It looks like the highest correlations with hospital performance in general are Median Household Income, % of 25+ with High School Degree, %White, and %Black. These are probably worth futher investigation. I am also interested if we controlled for Income, whether there would still be a significant correlation between hospital performance and the other three factors mentioned above.

```python
sns.set(font_scale=1.5)
pairplot = sns.pairplot(df)
```
