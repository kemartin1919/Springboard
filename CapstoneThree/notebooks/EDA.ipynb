```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df_pitchers = pd.read_csv('~/Desktop/Springboard/CapstoneThree/data/final/pitchers.csv')
df_batters = pd.read_csv('~/Desktop/Springboard/CapstoneThree/data/final/batters.csv')
```

```python
df_pitchers.info()
```

```python
df_batters.info()
```

We can see roughly 1/3 of the eligible pitchers were nominated to the HoF

```python
df_pitchers['HoF_nomination'].value_counts()
```

```python
pitchers_dict = {'Nominated': 340, 'Not Nominated': 846}
pitchers_df = pd.DataFrame(pitchers_dict.items(), columns=['HoF', 'Num_Pitchers'])
sns.barplot(x='HoF', y='Num_Pitchers', data=pitchers_df);
```

Same goes for eligible batters

```python
df_batters['HoF_nomination'].value_counts()
```

```python
batters_dict = {'Nominated': 528, 'Not Nominated': 949}
batters_df = pd.DataFrame(batters_dict.items(), columns=['HoF', 'Num_Batters'])
sns.barplot(x='HoF', y='Num_Batters', data=batters_df);
```

Scatterplot where x-axis is hits and y-axis is runs, hue shows whether or not the player was nominated to the HoF. There are some outliers but generally the trend is pretty clear.

```python
sns.scatterplot(x='h_y', y='r_y', hue='HoF_nomination', data=df_batters);
```

Trend is less clear when looking at walks and strikeouts. It seems that HoF nominees tend to walk a lot, but strike outs might not matter as much.

```python
sns.scatterplot(x='bb_y', y='so_y', hue='HoF_nomination', data=df_batters);
```

Scatterplot of pitchers, where x is wins and y is average runs allowed. We can see that HoF nominees tend to have a lot of wins, runs allowed is less of an obvious indicator.

```python
sns.scatterplot(x='w', y='r_x', hue='HoF_nomination', data=df_pitchers);
```

Swarmplot of hits for batters not nominated vs batters that were nominated

```python
sns.swarmplot(data=df_batters,
            x='HoF_nomination',
            y='h_y').set_title('Figure 1')
plt.show()
```

Same data visualized with a boxplot

```python
sns.boxplot(data=df_batters,
            x='HoF_nomination',
            y='h_y')
plt.show()
```

Heatmap of correlations between the different batter features. We can see many features are highly correlated, including average games played, at bats, runs, hits, doubles, RBIs, walks, and total games played

```python
sns.set(font_scale=3)
_ = plt.figure(figsize=(50,30))
sns.heatmap(df_batters.corr(), annot=True, annot_kws={"size": 30});
```

Heatmap of correlations between the different pitcher features. We can see many features are highly correlated, including wins, losses, games started, complete games, shutouts, outs pitched, hits given up, earned runs, homeruns given up, strikeouts, and walks.

```python
sns.set(font_scale=3)
_ = plt.figure(figsize=(50,30))
sns.heatmap(df_pitchers.corr(), annot=True, annot_kws={"size": 30});
```
