We want to know which offensive statistics are most important/indicative of success. In this case, success would mean more runs, wins, and playoff appearances. More specifically, we want to compare BA, OBP, and SLG percentages to try to determine which is the best indicator of success (and therefore, which should team management focus on increasing).

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('mlb_dataset.csv')
```

```python
df['RunDifferential'] = df['RS'] - df['RA']
```

```python
df.head()
```

```python
playoffs_ba = df.loc[df['Playoffs'] == 1, 'BA'].mean()
no_playoffs_ba = df.loc[df['Playoffs'] == 0, 'BA'].mean()
difference_ba = playoffs_ba / no_playoffs_ba - 1
print('Playoff teams have a ' + str(round(difference_ba * 100, 2)) + '% higher BA than non-playoff teams')
```

```python
playoffs_obp = df.loc[df['Playoffs'] == 1, 'OBP'].mean()
no_playoffs_obp = df.loc[df['Playoffs'] == 0, 'OBP'].mean()
difference_obp = playoffs_obp / no_playoffs_obp - 1
print('Playoff teams have a ' + str(round(difference_obp * 100, 2)) + '% higher OBP than non-playoff teams')
```

```python
playoffs_slg = df.loc[df['Playoffs'] == 1, 'SLG'].mean()
no_playoffs_slg = df.loc[df['Playoffs'] == 0, 'SLG'].mean()
difference_slg = playoffs_slg / no_playoffs_slg - 1
print('Playoff teams have a ' + str(round(difference_slg * 100, 2)) + '% higher SLG than non-playoff teams')
```

```python
Offense_dict = {'BA_increase': 3.65, 'OBP_increase': 4.23, 'SLG_increase': 6.91}
Offense_df = pd.DataFrame(Offense_dict.items(), columns=['Stat', '%Increase'])
sns.barplot(x='Stat', y='%Increase', data=Offense_df);
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

y=df['Playoffs']
X=df[['OBP', 'SLG', 'BA']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

scaler_x = MinMaxScaler((-1,1))
X_train = scaler_x.fit_transform(X_train)
X_test = scaler_x.transform(X_test)
```

Random Forest Classifier:

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)

y_predict_rf = rf.predict(X_test)
```

```python
from sklearn.model_selection import cross_val_score

cv_scores_test= cross_val_score(rf,X_test,y_test,cv=5,scoring='roc_auc')
cv_scores_train= cross_val_score(rf,X_train,y_train,cv=5,scoring='roc_auc')

cv_scores_rf_test= cv_scores_test.mean()
cv_scores_rf_train= cv_scores_train.mean()

print ('Mean cross validation test score: ' +str(cv_scores_rf_test))
print ('Mean cross validation train score: ' +str(cv_scores_rf_train))
```

```python
features = X
importances = list(rf.feature_importances_)
imp=np.sort(importances)
tab=pd.DataFrame(list(zip(X,imp)),columns =['Features', 'Importance scores']) 
print(tab)

#Visual comparison:
plt.figure(figsize=(10,5))
#indices = np.argsort(importances)
index = [0,1,2]
_=plt.barh(index,imp, align='center',color='b')
plt.xlabel('Relative Importance',fontsize=15)
plt.ylabel('Features',fontsize=15)
plt.yticks(index, features)
plt.title('Feature Importances for Random Forest classifier model',fontsize=15)
plt.show()
```

Gradient Boosting Classifier:

```python
from sklearn.ensemble import GradientBoostingClassifier
gbc = GradientBoostingClassifier(random_state=42)
gbc.fit(X_train, y_train)

y_predict_gbc = gbc.predict(X_test)
```

```python
cv_scores_test= cross_val_score(gbc,X_test,y_test,cv=5,scoring='roc_auc')
cv_scores_train= cross_val_score(gbc,X_train,y_train,cv=5,scoring='roc_auc')

cv_scores_gbc_test= cv_scores_test.mean()
cv_scores_gbc_train= cv_scores_train.mean()

print ('Mean cross validation test score: ' +str(cv_scores_gbc_test))
print ('Mean cross validation train score: ' +str(cv_scores_gbc_train))
```

```python
features = X
importances = list(gbc.feature_importances_)
imp=np.sort(importances)
tab=pd.DataFrame(list(zip(X,imp)),columns =['Features', 'Importance scores']) 
print(tab)

#visually:
plt.figure(figsize=(10,5))
#indices = np.argsort(importances)
index = [0,1,2]
_=plt.barh(index,imp, align='center',color='gold',edgecolor='k')
plt.xlabel('Relative Importance',fontsize=15)
plt.ylabel('Features',fontsize=15)
plt.yticks(index, features)
plt.title('Feature Importances for Gradient boost Classifier model',fontsize=15)
plt.show()
```
